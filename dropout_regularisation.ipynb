```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow
from tensorflow import keras
import seaborn as sns
```

```python
df=pd.read_csv('/content/sonar_dataset.csv',header=None)
df.head()
```

```python
X=df.drop(60,axis=1)
Y=df[60]
```

```python
X
```

```python
Y
```

```python
Y=pd.get_dummies(Y,drop_first=True)
#R-->1  and M-->0
```

```python
Y.head()
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import  train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)
```

```python
X_train.shape
```

```python
X_test.shape
```

**Model without Dropout Layer**

```python
model=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,epochs=100,batch_size=8)
```

```python
model.evaluate(X_test,Y_test)
```

**Training Accuracy >>> Test Accuracy**

```python
y_pred = model.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
Y_test[:10]
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(Y_test,y_pred))
```

```python
model_dropout=keras.Sequential([
    keras.layers.Dense(60,input_dim=60,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(15,activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1,activation='sigmoid'),


])
```

```python
model_dropout.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model_dropout.fit(X_train,Y_train,epochs=100,batch_size=8)
```

```python
model_dropout.evaluate(X_test,Y_test)
```

Training Accuracy is still good but Test Accuracy Improved

```python
y_pred = model_dropout.predict(X_test).reshape(-1)
print(y_pred[:10])

# round the values to nearest integer ie 0 or 1
y_pred = np.round(y_pred)
print(y_pred[:10])
```

```python
Y_test[:10]
```

```python
print(classification_report(Y_test,y_pred))
```
